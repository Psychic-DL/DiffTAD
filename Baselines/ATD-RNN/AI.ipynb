```python
%%cython

from libc cimport math

cdef double MIN_LON = -9 
cdef double MAX_LON = -8 
cdef double MIN_LAT = 41
cdef double MAX_LAT = 42
cdef int NUM_LONS = 840
cdef int NUM_LATS = 1100
cdef double PI = 3.1415926

cdef double DELTA_LON = (MAX_LON - MIN_LON) / NUM_LONS
cdef double DELTA_LAT = (MAX_LAT - MIN_LAT) / NUM_LATS

def get_lon_lat(int area_id):
    cdef int i = area_id // NUM_LONS
    cdef int j = area_id % NUM_LONS
    cdef double lon = j * DELTA_LON + MIN_LON
    cdef double lat = i * DELTA_LAT + MIN_LAT
    return (lon, lat)

def rad(double d):
    return d * PI / 180.0

def get_distance(double lat1, double lon1, double lat2, double lon2):
    cdef double radlat1 = rad(lat1)
    cdef double radlat2 = rad(lat2)
    cdef double a = radlat1 - radlat2
    cdef double b = rad(lon1) - rad(lon2)
    cdef double s = 2 * math.asin(math.sqrt(math.pow(math.sin(a/2), 2)\
                                  + math.cos(radlat1) * math.cos(radlat2) * math.pow(math.sin(b/2), 2)))
    s = s * 6378.137
    return s*10000 / 10000

def get_distance_id(int id1, int id2):
    cdef double lon1, lat1
    cdef double lon2, lat2
    lon1, lat1 = get_lon_lat(id1)
    lon2, lat2 = get_lon_lat(id2)
    dis = get_distance(lat1, lon1, lat2, lon2)
    return dis

def cal_anomaly_list(M, t=0.02):
    n = len(M)
    distance_array = []
    dis = [[0]*n for _ in range(n)]
    for i in range(len(M)):
        for j in range(i, len(M)):
            tmp_dis = get_distance_id(M[i], M[j])
            if tmp_dis>1e-12:
                distance_array.append(tmp_dis)
            else:
                tmp_dis = 1e-12
            dis[i][j], dis[j][i] = tmp_dis, tmp_dis
    distance_array.sort()
    print("distance_array sorted.")
    N = len(distance_array)
    d_c = distance_array[int(N * t)]
    print("N: ", N)
    print("d_c: ", d_c)
    density = []
    for i in range(n):
        tmp = 0
        for j in range(n):
            if i==j:
                continue
            tmp += math.exp(-(dis[i][j]/d_c)**2)
        density.append(tmp)
    AIList = []
    for i in range(n):
        tmp = 0
        for j in range(n):
            if i == j:
                continue
            if density[j]<1e-32:
                density[j] += 1e-32
            tmp += density[i] / density[j]
        tmp /= n
        AIList.append(tmp)
    AIList.sort()
    divFactorList = []
    for i in range(len(AIList)-1):
        divFactor = (AIList[i+1] - AIList[i]) / AIList[i]
        divFactorList.append(divFactor)
    
    avgDivFactor = sum(divFactorList) / len(divFactorList)
    sumDiv = 0
    for value in divFactorList:
        sumDiv += value
    divSize = len(divFactorList)
    avgDivFactor = sumDiv / divSize
    print("sumDiv: ", sumDiv)
    print("divSize: ", divSize)
    print("avgDivFactor: ", avgDivFactor)
    breakPoint = None
    anomalyList = []
    for i in range(len(AIList)-1):
        divFactor = (AIList[i+1] - AIList[i]) / AIList[i]
        # print(divFactor)
        if divFactor > avgDivFactor:
            anomalyList.append(M[i+1])
    return anomalyList
```

```python
lines = open("./output data/Porto/rnnTest7.txt").readlines()
MD = [list( map(int, line.split(" ")[2:])) for line in lines]
y_test = [1 - int(line.split(" ")[0]) for line in lines]
y_pred = []
for i in range(len(MD)):
    anomalyList = cal_anomaly_list(MD[i])
    if (len(anomalyList)) / len(MD[i]) > 0.05:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import (classification_report,
                            accuracy_score,
                            f1_score,
                            recall_score,
                            precision_score
                            )
```

```python
print("acc: ", accuracy_score(y_test, y_pred))
print("precision: ", precision_score(y_test, y_pred))
print("recall: ", recall_score(y_test, y_pred))
print("f1: ", f1_score(y_test, y_pred))
```
